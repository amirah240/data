````python
#your code goes here

#import the needed library
import pandas as pd

#define the dataframe here
iris = pd.read_csv('https://raw.githubusercontent.com/DSomair/Masar_DS/main/Datasets/DVQuiz/Iris.csv')


#check if you get the data
iris
````

````python
# The first way we can plot things is using the .plot extension from Pandas dataframes
# We'll use this to make a scatterplot of the Iris features.

#can you make a scatter plot?
import matplotlib.pyplot as plt

iris.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm")
````

````python
# One piece of information missing in the plots above is what species each plant is
# We'll use seaborn's FacetGrid to color the scatterplot by species

import seaborn as sns

import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'iris' is your DataFrame containing the Iris dataset

# Create a FacetGrid with the 'iris' DataFrame, specifying the 'hue' as the "Species" column
# 'size' parameter is deprecated; use 'height' instead
g = sns.FacetGrid(iris, hue="Species", height=5)

# Use the FacetGrid to map a scatter plot, plotting "SepalLengthCm" against "SepalWidthCm"
g.map(plt.scatter, "SepalLengthCm", "SepalWidthCm").add_legend()

# Show the plot
plt.show()
````

````python
# We can look at an individual feature in Seaborn through a boxplot
sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
````

````python
# A violin plot combines the benefits of the previous two plots and simplifies them
# Denser regions of the data are fatter, and sparser thiner in a violin plot
sns.violinplot(x="Species", y="PetalLengthCm", data=iris)
````

````python
# A final seaborn plot useful for looking at univariate relations is the kdeplot,
# which creates and visualizes a kernel density estimate of the underlying feature
sns.FacetGrid(iris, hue="Species") \
   .map(sns.kdeplot, "PetalLengthCm") \
   .add_legend()
````

````python
# Another useful seaborn plot is the pairplot, which shows the bivariate relation
# between each pair of features
#
# From the pairplot, we'll see that the Iris-setosa species is separataed from the other
# two across all feature combinations
sns.pairplot(iris.drop("Id", axis=1), hue="Species", size=3)
````

````python
# Now that we've covered seaborn, let's go back to some of the ones we can make with Pandas
# We can quickly make a boxplot with Pandas on each feature split out by species
iris.drop("Id", axis=1).boxplot(by="Species", figsize=(12, 6))
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the Iris dataset
iris = sns.load_dataset('iris')

# 1. Scatter plot x="SepalLengthCm", y="SepalWidthCm"
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', data=iris)
plt.title('Scatter plot of Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()

# 2. Scatter plot x="SepalLengthCm", y="SepalWidthCm" (Different color by species)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.title('Scatter plot of Sepal Length vs Sepal Width (colored by species)')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.show()

#3. boxplot for Petal Length for each species
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='petal_length', data=iris)
plt.title('Boxplot of Petal Length for each species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()

# 4. KDE plot for the Petal Length
plt.figure(figsize=(10, 6))
sns.kdeplot(data=iris, x='petal_length', hue='species', fill=True)
plt.title('KDE plot for Petal Length')
plt.xlabel('Petal Length (cm)')
plt.show()

# 5. Pairplot for all the variables (drop "Id") and make "hue" for each species
sns.pairplot(iris, hue='species', diag_kind='kde')
plt.show()
````
